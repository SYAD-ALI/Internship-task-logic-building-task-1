## Task 1 – User Login Check

```python
# Task 1 – User Login Check
def check_login():
  #stored credentials
  stored_username="syadali"
  stored_password="ali@1234"

  #Taking input from the user
  entered_username=input("Enter your username: ")
  entered_password=input("Enter your password: ")

  #verify
  if entered_username==stored_username and entered_password==stored_password:
    print("Login successful!")
  else:
    print("Login failed. Please check your credentials.")

#calling the function
check_login()
```

## Task 2 – Pass / Fail Analyzer

```python
#task 2 pass/fail analyzer
marks=[45,78,90,33,60,20,32]

totle_students=len(marks)
pass_students=0
fail_student=0

#assuming passing marks is 33
for mark in marks:
    if mark>=33:
        pass_students+=1
    else:
        fail_student+=1

print("total students:",totle_students)
print("pass students:",pass_students)
print("fail students:",fail_student)
```

## Task 3 – Simple Data Cleaner

```python
#Task 3 – Simple Data Cleaner
names=["Alice","bob","CHARLIE"]
cleaned_names=[]

for name in names:
  trimmed_name=name.strip()
  formatted_name=trimmed_name.lower()
  cleaned_names.append(formatted_name)

print("original names",names)
print("cleaned names",cleaned_names)

```

## Task 5 – Error Message Detector

```python
# Task 5 – Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("System Logs:", logs)
print("Total ERROR messages:", error_count)
```
